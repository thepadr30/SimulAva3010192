```python
import matplotlib.pyplot as plt
from IPython.display import clear_output, display


%matplotlib inline
# %matplotlib notebook
```

```python
def plano_cartesiano_jupyter():
    puntos_x = []
    puntos_y = []

    print("--- Modo Jupyter: Plano [-1, 1] ---")
    print("Escribe 'salir' para terminar.")

    while True:
        # 1. Creamos la figura en cada iteración para que se refresque
        fig, ax = plt.subplots(figsize=(6, 6))

        # Configuración fija del plano
        ax.set_xlim(-1.1, 1.1)
        ax.set_ylim(-1.1, 1.1)
        ax.axhline(0, color='blue', lw=1.5)
        ax.axvline(0, color='blue', lw=1.5)
        ax.grid(True, linestyle='--', alpha=0.6)
        ax.set_title(f"Puntos graficados: {len(puntos_x)}")
        ax.set_xlabel("Eje X")
        ax.set_ylabel("Eje Y")

        # 2. Dibujamos los puntos acumulados
        if puntos_x:
            ax.scatter(puntos_x, puntos_y, color='red', s=60, edgecolors='white', zorder=3)
            # Dibujar el último punto con una etiqueta para resaltar
            ax.text(puntos_x[-1] + 0.05, puntos_y[-1] + 0.05,
                    f"({puntos_x[-1]}, {puntos_y[-1]})", color='orange', fontsize=10)

        # 3. TRUCO PARA JUPYTER: Limpiar la celda y mostrar el nuevo gráfico
        clear_output(wait=True)
        display(fig)
        plt.close(fig) # Cerramos la figura para no saturar la memoria

        # 4. Pedir datos
        try:
            x_in = input("Introduce X (-1 a 1) o 'salir': ")
            if x_in.lower() == 'salir': break

            y_in = input("Introduce Y (-1 a 1) o 'salir': ")
            if y_in.lower() == 'salir': break

            x = float(x_in)
            y = float(y_in)

            if -1 <= x <= 1 and -1 <= y <= 1:
                puntos_x.append(x)
                puntos_y.append(y)
            else:
                print("⚠️ El punto debe estar entre -1 y 1. Intenta de nuevo.")
                input("Presiona Enter para continuar...")

        except ValueError:
            print("❌ Entrada no válida. Usa números.")
            input("Presiona Enter para continuar...")

    print("✅ Proceso terminado.")
```

```python
plano_cartesiano_jupyter()
```
